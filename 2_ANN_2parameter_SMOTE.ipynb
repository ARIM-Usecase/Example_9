### ライブラリのインポート
コード実行で必要なpythonのライブラリをimport文でロードします。

```python
# 汎用ライブラリ
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import itertools

# SMOTE
from imblearn.over_sampling import SMOTE

# 機械学習モデル
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense
from keras.optimizers import Adam

import warnings
warnings.filterwarnings('ignore')
```

# 1. データの準備
### サンプルファイルの読み込み
1と同じくdata.csvファイルをDataFrameとして読み込み、その結果をdfという変数に格納します。

```python
# CSVファイルからデータを読み込む
df = pd.read_csv('./data/data.csv')
df
```

# 2. 機械学習モデルの構築

```python
# 説明変数と目的変数の用意
X = df.iloc [:, :2] 
y = df.iloc [:, -1]
```

### Kerasによるモデル構築
本セクションでは、KerasのSequential APIを用いて人工ニューラルネットワーク（ANN）の構築します。Sequential APIは、深層学習モデルの層を順次追加することで、シンプルかつ直感的にモデルを設計できる手法です。

モデル内の各層では活性化関数（Activation Function）を設定する必要があり、ここでは代表的な活性化関数であるReLU（Rectified Linear Unit）を用います。ReLUは、非線形性をモデルに導入し、計算効率と学習の収束性を高める特性を持つため、深層学習において広く採用されています。

```python
# モデルの構築
model = Sequential([
    Input(shape=(2,)),  # 入力形状を指定
    Dense(64, activation='relu'),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(1, activation='linear')
])
```

【解説】モデルの各部分の役割

- **入力層**  
  最初に、モデルが受け取るデータの形を決めます。この例では、`shape=(2,)`と書かれているため、2つの特徴（例：重さや高さなど）を持つデータが入ってくることを意味しています。

- **隠れ層**  
  `Dense`という命令は、「全結合層」と呼ばれる層を作るためのものです。全結合層とは、各ニューロンが前の層のすべてのニューロンとつながっている層のことです。この例には3つの隠れ層があり、それぞれ次のように設定されています：

  - `Dense(64, activation='relu')`: 最初の隠れ層で、64個のニューロンを持っています。計算方法（活性化関数）は`relu`です。
  - `Dense(128, activation='relu')`: 次に、128個のニューロンを持つ隠れ層です。ここでも計算方法は`relu`です。
  - `Dense(64, activation='relu')`: 最後の隠れ層で、64個のニューロンがあり、計算方法は同じく`relu`です。

- **出力層**  
  `Dense(1, activation='linear')`で、1つの出力を持つ層を作ります。この層は最終結果を出すためのものです。`linear`という計算方法を使っているので、出力には特に変換がかかりません。

### モデルの設定（compileメソッド）
モデルの学習方法を決めるために、compileメソッドで損失関数や最適化アルゴリズムを設定します。

```python
model.compile(loss='mse', optimizer=Adam(learning_rate=0.0001))
```

【解説】compileメソッド

- **loss='mse'（損失関数）**  
  損失関数として「mse」（平均二乗誤差）を使っています。この関数は、モデルの予測値と実際の値の差の二乗をとり、その平均を計算して予測の精度を評価します。特に回帰問題では、この誤差を最小にすることで、予測の精度を向上させます。`mse`の値が小さくなるようにモデルが学習するため、誤差が小さくなるほど良い予測ができていることになります。

- **optimizer=Adam(learning_rate=0.0001)（最適化アルゴリズム）**  
  モデルを効率よく学習させるために、最適化アルゴリズムとして「Adam」を選んでいます。Adamは、通常の勾配降下法（重みを少しずつ調整する方法）よりも計算が早く、安定しているのが特徴です。  
  - **learning_rate=0.0001**：学習率はモデルの重みをどれくらい更新するかを決めるパラメータで、ここでは0.0001に設定しています。この値が大きすぎると不安定な学習になる可能性があり、小さすぎると学習が遅くなります。適切な値を選ぶことで、効率よく最適なモデルを作ることができます。

### モデルの学習（fitメソッド）
fitメソッドを使って、モデルにデータを与え、学習させます。このコードは、100回にわたって入力データ（X）と正解ラベル（y）をもとにモデルを学習し、モデルの予測精度を少しずつ向上させることを目指します。(注意：CPUのPCで約10分の処理時間がかかります)

```python
%%time
model.fit(X, y, epochs=100, batch_size=3)
```

【解説】fitメソッド  

- **X, y**：入力データとその正解ラベル（ターゲット）を指定します。ここで、`X`が特徴量（入力データ）、`y`がそれに対応する予測目標（正解データ）です。

- **epochs=100**：エポック数を指定します。1エポックは、モデルがすべての訓練データを1回通して学習することを指します。ここでは100回、すべてのデータを使って学習を繰り返します。エポック数が多いほど、モデルはデータに関する情報を多く学習できますが、多すぎると過学習（データに対して過度に適合すること）のリスクがあります。

- **batch_size=3**：バッチサイズは、一度にモデルに与えるデータの数を指定します。ここでは3つのデータポイントごとにモデルが学習を行います。バッチサイズが小さいと、モデルの重みが頻繁に更新され、学習が不安定になる場合がありますが、データが少ない場合には有効です。

# 3.予測モデル
水素と窒素の比率（ratio）によって、予測と実測のアンモニア濃度がどのように変化するかをプロットします。ここでは、アンモニア濃度と圧力の関係をグラフで視覚化するための関数 plot_predict()を事前に作成します。

```python
# グラフ描画用関数
def plot_predict(ratio, exp_points):
        
    pressures = np.arange(0, 81, 1)
    inputs = np.array([[p, ratio] for p in pressures])
    
    predictions = model.predict(inputs)

    plt.plot(pressures, predictions, label=f'Ratio={ratio}')
    plt.scatter(*zip(*exp_points))
    
    plt.xlabel('Pressure (psig)')
    plt.ylabel('NH3 concentration (%)')
    plt.legend()
```

【引数の説明】
* ratio：水素と窒素の比率を渡します。　異なる比率でのアンモニア濃度予測が可能になります。  
* exp_points：実際のデータ点（圧力とアンモニア濃度のペア）をリスト形式で渡します。　例として、[(0, 0.4), (20, 0.63), ...] のように記述します。  
* label：プロットの凡例に表示されるラベルです。凡例を示すために使用します。

### 作図
次のように引数を渡すことで、異なるratioに対するプロットが可能です。  

例）　plot_predict(0.5, [(0, 0.4), (20, 0.63), (40, 0.99), (80, 1.41)], 'Ratio=0.5')

ここでは、水素と窒素の比率として５つのケース（0.5, 1,2,3,5）について作図します。

```python
# プロットサイズを指定
plt.figure(figsize=(6, 4.5))

# 各水素：窒素比でのプロット
plot_predict(0.5, [(0, 0.4), (20, 0.63), (40, 0.99), (80, 1.41)])
plot_predict(1, [(0, 0.57), (20, 0.95), (40, 1.24), (60, 1.43), (75, 1.55), (80, 1.57)])
plot_predict(2, [(0, 0.6), (20, 0.96), (40, 1.16), (60, 1.27), (75, 1.33), (80, 1.35)])
plot_predict(3, [(0, 0.56), (20, 0.85), (40, 0.89), (80, 1.06)])
plot_predict(5, [(0, 0.48), (20, 0.63), (40, 0.63), (80, 0.69)])

plt.show()
```

上記の出力図から、人工ニューラルネットワーク（ANN）による予測モデルでは、任意の水素と窒素の比率において適切なフィッティングを達成できていないことが確認できます。

このようにスモールデータ（データ数が限られている状況）における機械学習モデルの適用には限界があります。スモールデータでは、モデルを一般化するために必要な学習データが不足するため、学習結果が不安定となり、予測精度の低下を引き起こしています。

このような課題に対応するため、データの拡充を目的とした手法として、SMOTE（Synthetic Minority Over-sampling Technique）を適用した検討を次に進めます。

SMOTEは、既存のデータを基に新たな合成データポイントを生成し、データセットの偏りや不足を補完する手法です。このアプローチにより、データ不足による学習性能の低下を改善できるかを見てみることにしましょう。

# 4. SMOTEによるデータ生成

```python
# SMOTEによるデータ生成
smote = SMOTE(k_neighbors=1, random_state=0)
```

```python
# インデックスの組み合わせを生成してSMOTEに使用
index_list = list(range(df.shape[0]))
ind_set = list(itertools.combinations(index_list, 2))
```

```python
# インデックスの組み合わせにおける総当たりでのSMOTE計算
smote_data = []

for item in ind_set:
    indices = list(item)
    y_smote = np.zeros(df.shape[0])
    y_smote[indices] = 1
    df_resampled, _ = smote.fit_resample(df, y_smote)
    smote_data.append(df_resampled)
```

```python
# NumPy配列に変換してリシェイプ
smote_data = np.array(smote_data).reshape(-1, 3)
smote_data
```

```python
smote_data.shape
```

# 5. SMOTEデータへの機械学習モデルの適用

### データセットの分割（Data set splitting）

```python
# 説明変数と目的変数の用意
X_smote = smote_data[:, :2] 
y_smote = smote_data[:, -1]
```

### Kerasによるモデル構築

```python
# モデルの構築
model_2 = Sequential([
    Input(shape=(2,)),  # 入力形状を指定
    Dense(64, activation='relu'),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(1, activation='linear')
])
```

### モデルの設定（compileメソッド）

```python
model_2.compile(loss='mse', optimizer=Adam(learning_rate=0.0001))
```

```python
%%time
model_2.fit(X_smote, y_smote, epochs=100, batch_size=3)
```

# 6.予測モデル（SMOTE）
水素と窒素の比率（ratio）によって、予測と実測のアンモニア濃度がどのように変化するかをプロットします。

```python
# グラフ描画用関数
def plot_predict(ratio, exp_points):
        
    pressures = np.arange(0, 81, 1)
    inputs = np.array([[p, ratio] for p in pressures])
    
    predictions = model_2.predict(inputs)

    plt.plot(pressures, predictions, label=f'Ratio={ratio}')
    plt.scatter(*zip(*exp_points))
    
    plt.xlabel('Pressure (psig)')
    plt.ylabel('NH3 concentration (%)')
    plt.legend()
```

```python
# プロットサイズを指定
plt.figure(figsize=(6, 4.5))

# 各水素：窒素比でのプロット
plot_predict(0.5, [(0, 0.4), (20, 0.63), (40, 0.99), (80, 1.41)])
plot_predict(1, [(0, 0.57), (20, 0.95), (40, 1.24), (60, 1.43), (75, 1.55), (80, 1.57)])
plot_predict(2, [(0, 0.6), (20, 0.96), (40, 1.16), (60, 1.27), (75, 1.33), (80, 1.35)])
plot_predict(3, [(0, 0.56), (20, 0.85), (40, 0.89), (80, 1.06)])
plot_predict(5, [(0, 0.48), (20, 0.63), (40, 0.63), (80, 0.69)])

plt.show()
```
